```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

transactions = []

data = "german_encoded.csv"
df = pd.read_csv(data, sep=",", decimal=".")
```

```python
from pymining import itemmining
relim_input = itemmining.get_relim_input(df)
item_sets = itemmining.relim(relim_input, min_support=0.2)
```

```python
item_sets
```

```python
from pymining import assocrules
rules = assocrules.mine_assoc_rules(item_sets, min_support=0.2, min_confidence=0.5)
for rule in rules:
    print(rule)
```
