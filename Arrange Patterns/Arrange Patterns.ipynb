```python
import pandas as pd
from ast import literal_eval
import numpy as np
```

```python
path = './datasets'
path_res1='%s/bank2_final_women.csv' % (path)
```

```python
data1 = pd.read_csv(path_res1, sep=",", header=0, converters={"sequence": literal_eval})
```

```python
dataset_size = len(data1.values.tolist())
dataset_size
```

```python
path = './patterns'
name = 'result_tx_women_ps_supp_175761.csv'
path_res2='%s/%s' % (path, name)
```

```python
data = pd.read_csv(path_res2, sep=",", header=0, converters={"patterns": literal_eval})
data
```

```python
patterns = list(data['patterns'])
```

```python
import numpy as np
len_seq = []
for s in patterns:
    len_seq.append(len(s))
seq_len = np.array(len_seq)
seq_len
```

```python
max_itemsets = max(seq_len)
max_itemsets
```

```python
data_list = data.values.tolist()
data_list
```

```python
labels = []
for i in range(1,max_itemsets + 1):
    labels.append("IS" + str(i))
labels.append("supp")
labels.append("%supp")
labels      
```

```python
df = pd.DataFrame(columns=labels)
df
```

```python
list_series = []
for r in data_list:
    row_l = r[0][:]
    while len(row_l) < max_itemsets:
        row_l.append(' ')
    row_l.append(r[1])
    row_l.append(round((float(r[1])/float(dataset_size)),2))
    list_series.append(pd.Series(row_l, index=df.columns))
```

```python
df2 = df.append(list_series , ignore_index=True)
```

```python
df2
```

```python
df3 = pd.DataFrame(columns=labels)
```

```python
df3
```

```python
list_series2 = []
for r in data_list:
    row_l = []
    for itemset in r[0]:
        if type(itemset) == list:
            item  = "< "
            for item in itemset:
                item =+ str(item) + " "
            item += ">"
        else:
            row_l.append("< " + str(itemset) + " >")
        
    while len(row_l) < max_itemsets:
        row_l.append(" ")
        
    row_l.append(r[1])
    row_l.append(round((float(r[1])/float(dataset_size)),2))
    list_series2.append(pd.Series(row_l, index=df.columns))
```

```python
df4 = df3.append(list_series2 , ignore_index=True)
```
